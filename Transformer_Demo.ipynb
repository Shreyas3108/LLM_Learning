## Tokenizer

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I am learning from HuggingFace.",
        "I hate this so much!",
    ]
)
```

```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
inp = [
        "I am learning from HuggingFace.",
        "I hate this so much!",
    ]
```

```python
inputs = tokenizer(inp , padding = True , truncation = True , return_tensors = "pt")
```

```python
inputs
```

## Going through the model

```python
from transformers import AutoModel

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(checkpoint)
```

```python
model.transformer
```

```python
outputs = model(**inputs)
```

```python
print(outputs.last_hidden_state.shape)
```

```python
import torch

predictions = torch.nn.functional.softmax(outputs['last_hidden_state'], dim=-1)
print(predictions)
```

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
```

```python
print(outputs.logits.shape)
```

```python
print(outputs.logits)
```

```python
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

```python
model.config.id2label
```

```python
import pandas as pd
import numpy as np
pd.DataFrame(predictions.detach().numpy() , columns = ['negative' , 'postive'] , index = [0,1])
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."

tokens = tokenizer.tokenize(sequence)
ids = tokenizer.convert_tokens_to_ids(tokens)

input_ids = torch.tensor([ids])
print("Input IDs:", input_ids)

output = model(input_ids)
print("Logits:", output.logits)
```
